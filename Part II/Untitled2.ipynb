## Weather Dataset

```python
# Load weather data
weather_data = pd.read_csv('weather_Station_data.csv')
```

```python
# update the column name 'Date/Time' to 'Date' to be consistent with other datasets
weather_data.rename(columns={'Date/Time': 'Date'}, inplace=True)
```

```python
# Calculate monthly means for 'Mean Max Temp (°C)' and 'Mean Min Temp (°C)'
monthly_means = weather_data.groupby(['Month'])[['Mean Max Temp (°C)']].mean()

# Use these means to fill missing values
weather_data['Mean Max Temp (°C)'] = weather_data['Mean Max Temp (°C)'].fillna(weather_data['Month'].map(monthly_means['Mean Max Temp (°C)']))
#weather_data['Mean Min Temp (°C)'] = weather_data['Mean Min Temp (°C)'].fillna(weather_data['Month'].map(monthly_means['Mean Min Temp (°C)']))
```

```python
# List of columns to fill missing values
columns_to_fill = ['Mean Min Temp (°C)', 'Mean Temp (°C)', 'Extr Max Temp (°C)',
                    'Extr Min Temp (°C)']

# Group by 'Month' and calculate mean for each column
monthly_means = weather_data.groupby(['Month'])[columns_to_fill].mean()

# Iterate over the columns and fill missing values with monthly means
for column in columns_to_fill:
    weather_data[column].fillna(weather_data['Month'].map(monthly_means[column]), inplace=True)

# Check the result
print(weather_data[columns_to_fill].isnull().sum())
```
