## Weather Dataset

```python
# Load weather data
weather_data = pd.read_csv('weather_Station_data.csv')
```

```python
# update the column name 'Date/Time' to 'Date' to be consistent with other datasets
weather_data.rename(columns={'Date/Time': 'Date'}, inplace=True)
```

```python
weather_data
```

```python
weather_data['Date'] = pd.to_datetime(weather_data['Date'])
```

```python
# Calculate monthly means for 'Mean Max Temp (°C)' and 'Mean Min Temp (°C)'
monthly_means = weather_data.groupby(['Month'])[['Mean Max Temp (°C)', 'Mean Min Temp (°C)']].mean()

# Use these means to fill missing values
weather_data['Mean Max Temp (°C)'] = weather_data['Mean Max Temp (°C)'].fillna(weather_data['Month'].map(monthly_means['Mean Max Temp (°C)']))
weather_data['Mean Min Temp (°C)'] = weather_data['Mean Min Temp (°C)'].fillna(weather_data['Month'].map(monthly_means['Mean Min Temp (°C)']))
```

```python
columns_to_fill_zero = ['Total Snow (cm)', 'Total Precip (mm)', 'Snow Grnd Last Day (cm)', 'Dir of Max Gust (10\'s deg)', 'Spd of Max Gust (km/h)']
weather_data[columns_to_fill_zero] = weather_data[columns_to_fill_zero].fillna(0)
```
